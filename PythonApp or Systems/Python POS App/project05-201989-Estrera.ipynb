```python
#Problem A: POS Machine, improved
import csv
defaulttrans = 'ID  Product  Price  Quantity  Total'

class Posmachine(): 
       
    def __init__(self):
        self.countint = 0
        self.countint1 = self.countint + 1
        self.countstring = str(self.countint)
        self.invoice = 'invoice'
        self.txt = '.txt'
        self.myid = []
        self.identity = []
        self.listid = []
        self.listq = []
        self.product = []
        self.total = []
        self.total1 = []
        self.reportfinal = []
        self.grandtotal = 0
        self.greattotal = 0
        self.final = 0
        self.sumofall = 0
        self.a = ' '
        self.file = ' '
        self.openfile = ' '
        self.callcount = 0
        self.reportcount = 0
        self.reportlist = []

       
    
    def setPriceList(self):
        try:
            file_path = str(entryfile.get())
            file1 = open(file_path,'r')
            self.file = file_path
            self.openfile = file1
            if file1:  
                statusfile.configure(text='File opened')
            else:
                statusfile.configure(text='File does not exist')
        except:
            print('Invalid')
        
               
            
    def printPriceList(self):
        try:        
            read = self.openfile.readlines()
            remove = [a.replace(',','  ') for a in read]
            yes = [pricelist.insert(INSERT, a) for a in remove] 
        except:
            pricelist.configure(text='No Price List set')
            
    def removePriceList(self):
        pricelist.delete('1.0',END)
        self.openfile.close()
                       
                
                        
    def addProductToTransaction(self):
        listid = [] #product IDs
        listq = [] #quantities
        listtotal = [] #Sum of all products in a transaction
        self.listid = listid
        self.listq = listq
        obtainid = str(identry.get())
        obtainquant = int(idquant.get())
        listid.append(obtainid)
        listq.append(obtainquant)
  
        if self.file == 'price_list1.csv':  #for csv
            if self.callcount == 0:
                translist.insert(INSERT,defaulttrans)
        
            with open(self.file,'r') as csvfile:
                reader = csv.DictReader(csvfile)
                comp1 = [] #list of prices
                product = [] #list of product 
                self.product = product
                for row in reader:
                    for id in self.listid:
                        if row['ID'] == id:
                            comp1.append(int(row['Price']))

                for id, quant in zip(comp1, self.listq):
                    product.append(id * quant)
                    
                
                grandtotal = sum(product) #sum of individual items
                self.grandtotal = grandtotal 
         
    
            with open(self.file,'r') as csvfile:
                reader = csv.DictReader(csvfile)
                for row in reader:
                    for ide, quant, prod in zip(self.listid,self.listq,self.product):
                        if row['ID'] == ide:
                            translist.insert(INSERT,'\n')
                            translist.insert(INSERT,row['ID'])
                            translist.insert(INSERT,'   ')
                            translist.insert(INSERT,row['Product'])
                            translist.insert(INSERT,'   ')
                            translist.insert(INSERT,row['Price'])
                            translist.insert(INSERT,'   ')
                            translist.insert(INSERT,quant)
                            translist.insert(INSERT,'   ')
                            translist.insert(INSERT,self.grandtotal)
                translist.insert(INSERT, '\n')
            self.total.append(self.grandtotal)
            
            
        
        elif self.file == 'price_list2.txt': #for txt
            if self.callcount == 0:
                translist.insert(INSERT,defaulttrans)
                
            with open(self.file,'r') as txtfile:
                id1 = [] # main ID 
                price = [] 
                name = []
                prices = [] # prices received
                finalprices = [] # prices with product
                read = txtfile.readlines()
             
                for line in read:
                    split = line.split('\t')
                    id1.append(split[0])
                    price.append(split[2])
                    name.append(split[1])
                    
                newprice = [a.replace('\n','') for a in price]  
                newprice.remove('Price')
                mainprice = [int(a) for a in newprice] # main prices   
                id1.remove('ID')
                name.remove('Product')
                
                for id in self.listid:
                    for a, b in zip(id1,mainprice):
                        if a == id:
                            prices.append(b)
                        
                for p, quant in zip(prices,self.listq):
                    finalprices.append(p*quant)
                                  
                greattotal = sum(finalprices)
                self.greattotal = greattotal #Sum of prices availed
                self.greatproduct = finalprices #Final prices of products
                self.identity = name
                self.myid = id1 #All IDs
                self.initialprices = prices #initial prices
                

                with open(self.file,'r') as txtfile:
                    for x, name in zip(self.myid,self.identity):
                        for ide, price, quant, product in zip(self.listid,self.initialprices,self.listq,self.greatproduct):
                            if ide == x:
                                translist.insert(INSERT,'\n')
                                translist.insert(INSERT,x)
                                translist.insert(INSERT,'  ')
                                translist.insert(INSERT, name)
                                translist.insert(INSERT,'  ')
                                translist.insert(INSERT,price)
                                translist.insert(INSERT,'  ')
                                translist.insert(INSERT, quant)
                                translist.insert(INSERT,'  ')
                                translist.insert(INSERT, product)
                                
                    translist.insert(INSERT,'\n') 
            self.total1.append(self.greattotal) 
            
        self.callcount +=1
        
               
         
                    
    def makeNewTransaction(self):
        self.savereport()
        translist.delete('1.0', END)
        invoice.configure(text=self.countint)
                    
                 
    def createSalesInvoice(self):
        self.countint += 1
        self.callcount = 0
        if self.file=='price_list1.csv':
            totality = sum(self.total)
            self.final = totality
            translist.insert(INSERT, '------------------------------------')
            translist.insert(INSERT, '\n')
            translist.insert(INSERT,'Grand Total')
            translist.insert(INSERT, '     ')                  
            translist.insert(INSERT, self.final)
            self.total.clear()
            with open(self.invoice+str(self.countint)+self.txt,'w') as k:
                getting = translist.get('1.0',END)
                print(getting,file=k)
            self.reportfinal.append(self.final)
        
                        
        elif self.file=='price_list2.txt':
            totality1 = sum(self.total1)
            self.final1 = totality1
            translist.insert(INSERT, '------------------------------------')
            translist.insert(INSERT, '\n')
            translist.insert(INSERT, 'Grand Total:')
            translist.insert(INSERT, '     ')
            translist.insert(INSERT, self.final1)
            self.total1.clear()
            with open(self.invoice+str(self.countint)+self.txt,'w') as writefile:
                getting1 = translist.get('1.0',END)
                print(getting1,file=writefile)
            self.reportfinal.append(self.final1)
                
    def savereport(self):
        translist.delete('end-1l','end')
        a = translist.get('2.0','end')
        self.a = a
        self.reportlist.append(self.a)
        
    def generatereport(self):
        sumofall = sum(self.reportfinal)
        self.sumofall = sumofall
        reportlist.insert(INSERT,'Sales Report')
        reportlist.insert(INSERT,'\n')
        reportlist.insert(INSERT,defaulttrans)
        reportlist.insert(INSERT,'\n')
        empty = ' '
        k = empty.join(self.reportlist)
        reportlist.insert(INSERT, k)
        reportlist.insert(INSERT,'\n')
        reportlist.insert(INSERT,'Grand Total:')
        reportlist.insert(INSERT,'    ')
        reportlist.insert(INSERT,self.sumofall)
       
    def clearreport(self):
        reportlist.delete('1.0',END)
            
    

        
        
```
